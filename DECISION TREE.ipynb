```python
# DECISION TREE :
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a Decision Tree Classifier
clf = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)

# Train the model
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Visualize the tree (text format)
tree_rules = export_text(clf, feature_names=iris.feature_names)
print(tree_rules)
```

```python
# SUPPORT VECTOR MACHINES :
from sklearn.svm import SVC
import numpy as np
import matplotlib.pyplot as plt

# Data points
X = np.array([[2, 3], [3, 4], [5, 6], [6, 7]])  # Features
y = np.array([0, 0, 1, 1])  # Labels (0 for Class A, 1 for Class B)

# Train SVM model
svm = SVC(kernel='linear', C=1)
svm.fit(X, y)

# Plotting the data points and hyperplane
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', s=100)
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create a grid to evaluate the model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = svm.decision_function(xy).reshape(XX.shape)

# Plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
plt.show()
```
