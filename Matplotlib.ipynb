```python
# (5) Horizontal bar plot.
import pandas as pd
import matplotlib.pyplot as plt
data={
    'month':['jan','feb','mar','apr'],
    'sales':[300,200,100,400]
}
frame=pd.DataFrame(data)
frame.plot(x="month",y="sales",kind='barh',color='green')
plt.title("Data")
plt.xlabel("sales")
plt.ylabel("months")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data={
    'month':['jan','feb','mar','apr'],
    'sales':[300,200,100,400]
}
c=['red' if m=="apr" else "green" for m in data['month']] # Giving condition
frame=pd.DataFrame(data)
frame.plot(x="month",y="sales",kind='barh',color=c)
plt.title("Data")
plt.xlabel("sales")
plt.ylabel("months")
plt.show()
```

```python
# (7) hexbin plot
import pandas as pd
import matplotlib.pyplot as plt
data={
    'month':[1,2,3,4],
    'sales':[300,200,100,400]
}
frame=pd.DataFrame(data)
frame.plot(x="month",y="sales",kind='hexbin',color='orange',gridsize=30) # kind = hexbin -> used for hexbin plot. It is mainly used in medical science.
plt.show()
```

```python
# (8) Histogram 
import numpy as np
import matplotlib.pyplot as plt
mean_score=70
st_dev=10
num_student=1000
exam_scores=np.random.normal(mean_score,st_dev,num_student)
plt.hist(exam_scores,bins=30,color='blue',edgecolor='red')
plt.title("Histogram of marks")
plt.show()
```
