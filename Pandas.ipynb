```python
# Time series
# resample()
import pandas as pd
import numpy as np

# Create a time-series DataFrame
date_rng = pd.date_range(start='2023-01-01', end='2023-01-10', freq='D')
data = {'date': date_rng, 'value': np.random.randint(0, 100, size=(len(date_rng)))}
df = pd.DataFrame(data)

# Set date as the index
df.set_index('date', inplace=True)

# Resample to 3-day frequency and compute the sum
df_resampled = df.resample('3D').sum()
df_resampled
```

```python
# Merging():
# The merge() function in pandas is used to combine two DataFrames based on one or more keys (columns or indexes). 
# It is similar to SQL joins and allows for a variety of join operations.
# By default inner join is shown:
import pandas as pd

df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Age': [24, 27, 30]})

result = pd.merge(df1, df2, on='ID')
print(result)
```

```python
# Left join:
import pandas as pd

df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Age': [24, 27, 30]})

result = pd.merge(df1, df2, on='ID',how="left")
print(result)
```

```python
# Right join:
import pandas as pd

df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Age': [24, 27, 30]})

result = pd.merge(df1, df2, on='ID',how="right")
print(result)
```

```python
# Outer join:
import pandas as pd

df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Age': [24, 27, 30]})

result = pd.merge(df1, df2, on='ID',how="outer")
print(result)
```

```python
# melt():
# The melt() function in pandas is used to transform a DataFrame from a wide format to a long format. 
# It unpivots the columns of a DataFrame into rows, making it particularly useful for reshaping data for analysis or visualization.
import pandas as pd

df = pd.DataFrame({
    'ID': [1, 2, 3],
    'Math': [85, 90, 95],
    'Science': [80, 85, 90]
})

result = pd.melt(df, id_vars=['ID'], value_vars=['Math', 'Science'], var_name='Subject', value_name='Score')
print(result)
```
